Using Corr() to see the correlation between the columns

```python
df.columns
df_corr = df[['popularity','duration_ms', 'danceability', 'energy',
             'key', 'loudness','mode', 'speechiness', 'acousticness', 'instrumentalness',
             'liveness','valence', 'tempo', 'time_signature']].corr()
df_corr
```

Plotting heatmap using corr() - this shows the correlation between the tables with the correlation coeffecient on the map. It's a Pearson's correlation

Darker plots show they are negatively correlated and lighter colors are positively correlated. Also Correlation point to itelf will perfectly correlated and will always be 1.

Notice in this heatmap - energy and loudness are closely correlated (.76 points) 

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr, annot=True, cmap='mako')
```

Using Seaborn pairplot - with acousticness and energy variables

First plot is just a basic one with all the values and you can see the scatterplot seems very crowded. 
Second plot is more refined with just 5000 samples.

```python
sns.pairplot(df[['acousticness','energy']], height =5, aspect= 1.5)
```

```python
df_sample = df[['acousticness', 'energy']].sample(n=5000)
sns.pairplot(df_sample, height=5, aspect = 1.5, diag_kind='kde')
```

```python
df['track_genre'].value_counts()
```

To show we can add more variables to pairplot. 

Here we have selected 5 different genres. 5000 samples. random_state=1 uses same set of samples everytime you run the visualization code. If you dont mention random_state=1, it will pick up different samples every time you run the code. 

```python
df_genres = df[df['track_genre'].isin(['pop', 'folk', 'emo', 'punk-rock', 'electronic'])]

df_sample = df_genres.sample(n=5000, random_state=1)

sns.pairplot(df_sample[['acousticness', 'energy', 'track_genre']], height=5, aspect = 1.5, diag_kind='kde', hue='track_genre')
```
